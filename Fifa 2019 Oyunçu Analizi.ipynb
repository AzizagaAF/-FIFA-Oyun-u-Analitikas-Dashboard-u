## Preprocessing (Oyunçu dəyərinin rəqəmə çevrilməsi)

##### Value sutununa baxmaq isteyirem. Sutunun ilk 10 reqemini nece gore bilerem?

```python
df['Value'].head(20)
```

##### Burda Euro isaresini cixaririq, M herfini 1000000-la, K herfini ise 1000-le evezleyirik

```python
def convert_value(val):
    if isinstance(val, str):
        val = val.replace('€', '') 
        if 'M' in val:
            return float(val.replace('M', '')) * 1000000 
        elif 'K' in val:
            return float(val.replace('K', '')) * 1000
        return float(val)
    return val

df['Value'] = df['Value'].apply(convert_value)
```

```python
df['Value'].head(20)
```

```python
df.describe()
```

##### Menim sutunumda null deyerler ola biler. Men isteyirem ki bunu 0 ile evez edim

```python
df['Value'] = df['Value'].fillna(0)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import MultipleLocator, FuncFormatter

fig, axes = plt.subplots(4, 2, figsize=(20, 35))
plt.subplots_adjust(hspace=0.4, wspace=0.3)

order = df['Position'].value_counts().index
ax1 = sns.countplot(ax=axes[0, 0], y='Position', data=df, order=order, palette='viridis')
for container in ax1.containers:
    ax1.bar_label(container, fmt='%d', padding=5, fontsize=10, fontweight='bold')
axes[0, 0].set_title('Oyunçuların Mövqelərinə Görə Sayı', fontsize=16, fontweight='bold')

ag = sns.histplot(ax=axes[0, 1], x=df['Age'], bins=20, color='teal', alpha=0.4)
for container in ag.containers:
    ag.bar_label(container, fmt='%d', padding=5, fontsize=10, color='red', fontweight='bold')
ax2_kde = axes[0, 1].twinx()
sns.kdeplot(df['Age'], fill=True, color="darkorange", linewidth=3, alpha=0.3, ax=ax2_kde)
axes[0, 1].axvline(df['Age'].mean(), color='red', linestyle='--')
axes[0, 1].set_title('Yaş Paylanması və Sıxlığı', fontsize=16, fontweight='bold')

foot_counts = df['Preferred Foot'].value_counts()
axes[1, 0].pie(foot_counts, labels=foot_counts.index, autopct='%1.1f%%', startangle=140, 
               colors=['lightblue', 'lightgreen'], explode=(0.05, 0), shadow=True,
               textprops={'fontsize': 14, 'weight': 'bold'})
axes[1, 0].set_title('Üstünlük Verdiyi Ayaqların Faiz Payı', fontsize=16, fontweight='bold')

ax4 = sns.barplot(ax=axes[1, 1], x='Position', y='Value', hue='Preferred Foot', data=df, palette='Spectral', ci=None)
for container in ax4.containers:
    labels = [f'{v.get_height()/1e6:.1f}M' if v.get_height() > 0 else '' for v in container]
    ax4.bar_label(container, labels=labels, fontsize=8, fontweight='bold', padding=3, rotation=90)
ax4.yaxis.set_major_formatter(FuncFormatter(lambda x, pos: f'€{x/1e6:.0f}M'))
axes[1, 1].tick_params(axis='x', rotation=45)
axes[1, 1].set_title('Mövqe və Ayağa Görə Orta Qiymət (€)', fontsize=16, fontweight='bold')

ax5 = sns.stripplot(ax=axes[2, 0], x='International Reputation', y='Potential', hue='Preferred Foot', 
                    data=df, jitter=True, dodge=True, alpha=0.6, palette={'Left': '#e74c3c', 'Right': '#3498db'})
axes[2, 0].yaxis.set_major_locator(MultipleLocator(5))
axes[2, 0].set_title('Reputasiya və Potensial (Stripplot)', fontsize=16, fontweight='bold')

sns.boxplot(ax=axes[2, 1], x='International Reputation', y='Potential', hue='Preferred Foot', 
            data=df, palette={'Left': '#e74c3c', 'Right': '#3498db'}, width=0.7)
axes[2, 1].set_title('Reputasiya və Potensial (Boxplot)', fontsize=16, fontweight='bold')

sns.lineplot(ax=axes[3, 0], x='Overall', y='Potential', hue='Preferred Foot', data=df, 
             palette={'Left': '#e74c3c', 'Right': '#3498db'}, linewidth=3)
axes[3, 0].grid(True, linestyle='--', alpha=0.5)
axes[3, 0].set_title('Güc (Overall) və Potensial Əlaqəsi', fontsize=16, fontweight='bold')

ax8 = sns.barplot(ax=axes[3, 1], x='Preferred Foot', y='Value', data=df, palette=['lightblue', 'lightgreen'], ci=None)
for container in ax8.containers:
    ax8.bar_label(container, fmt='€%.0f', label_type='edge', fontsize=12, fontweight='bold')
ax8.yaxis.set_major_formatter(FuncFormatter(lambda x, pos: f'€{x/1e6:.1f}M'))
axes[3, 1].set_title('Ayağa Görə Orta Oyunçu Qiyməti', fontsize=16, fontweight='bold')
plt.suptitle('FIFA Oyunçu Analitikası Dashboard-u', fontsize=24, fontweight='bold', y=0.92)
plt.savefig('fifa_dashboard.png', dpi=300, bbox_inches='tight')
plt.show()
```
